```python
from bs4 import BeautifulSoup
```

```python
def parse_html(text):
    soup = BeautifulSoup(text, 'html.parser')
    # Replace URLs with URL
    for link in soup.find_all('a', href = True):
        link.replace_with(link["href"])
    # Remove code, quotes, JS, styles
    for tag in soup.find_all(['code', 'blockquote', 'script', 'style']):
        tag.decompose()
    res = soup.get_text()
    return res
```

```python
frame['raw_text'] = frame['Body'].apply(lambda text: parse_html(text))
frame.head()
```

```python
def bylength(word1,word2):
    return len(word2)-len(word1)


def generate_examples(frame, topic):
    threshold = 2000
    groups = frame.groupby('Id')
    examples = []
    for group in groups:
        posts = group[1]['Body']
        long_posts = []
        short_posts = []
        for post in posts:
            if len(post) > threshold:
                long_posts.append(post)
            else:
                short_posts.append(post)
        if len(long_posts) == 0:
            continue
        long_posts = sorted(long_posts, key = len)
        long_posts.reverse()
        short_posts = sorted(short_posts, key = len)
        short_posts.reverse()
        
        count_long = len(long_posts)
        count_short = len(short_posts)
        if 2 * count_long <= count_short:
            for i in range(count_long):
                examples.append([long_posts[i], short_posts[i], short_posts[i + count_long]])
        else:
            num_triples = (count_long + count_short) // 3
            all_posts = long_posts + short_posts
            for i in range(num_triples):
                candidate = [all_posts[i], all_posts[i + num_triples], all_posts[i + 2 * num_triples]]
                if len(candidate[0]) > threshold:
                    examples.append(candidate)
    print(len(examples))
    for i in range(10):
        for example in examples[i]:
            print(example)
            print("=====================")
        print("***********************")
    return examples
```

```python
examples = generate_examples(frame, "Hello")
```

```python

df = pd.DataFrame(index = np.arange(0, len(examples)), columns = ['text', 'positions'])
curr = 0
for example in examples:
    text = example[0] + example[1]
    splits = [len(example[0]), len(example[0]) + len(example[1])]
    df.loc[curr] = {'text': "".join(example), 'positions' : splits}
    curr = curr + 1
df.head()
```
